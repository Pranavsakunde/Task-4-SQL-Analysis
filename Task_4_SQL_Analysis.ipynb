```python
!pip install pandasql

import pandas as pd
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

df.columns = df.columns.str.lower().str.strip()

print("Columns:", df.columns)

brand_col = None
price_col = None
ram_col = None

for col in df.columns:
    if "brand" in col:
        brand_col = col
    if "price" in col:
        price_col = col
    if "ram" in col:
        ram_col = col

print("Detected:", brand_col, price_col, ram_col)

df['brand'] = df[brand_col] if brand_col else "unknown"

df['price'] = pd.to_numeric(df[price_col], errors='coerce') if price_col else 0

if ram_col:
    df['ram'] = pd.to_numeric(df[ram_col], errors='coerce')
else:
    df['ram'] = None

df = df.dropna(subset=['price'])

print("\nPreview:")
print(df[['brand','price','ram']].head())


print("\nTop 5 Expensive:")
print(pysqldf("SELECT brand, price FROM df ORDER BY price DESC LIMIT 5"))

print("\nAverage Price by Brand:")
print(pysqldf("SELECT brand, AVG(price) FROM df GROUP BY brand"))

print("\nCount by Brand:")
print(pysqldf("SELECT brand, COUNT(*) FROM df GROUP BY brand"))

if df['ram'].notnull().any():
    print("\nRAM > 8GB:")
    print(pysqldf("SELECT brand, price, ram FROM df WHERE ram > 8"))
else:
    print("\nRAM column not found or empty — skipping this query")
```
